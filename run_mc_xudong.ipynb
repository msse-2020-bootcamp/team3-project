# Monte Carlo Simulation

Run Monte Carlo simulation from imported module.

```python
import os
import mcsim.monte_carlo as mc
```

```python
help(mc.run_simulation)
```

```python
# Set the initial parameters
reduced_temperature = 0.9
num_steps = 500
max_displacement = 0.1
cutoff = 3
freq = 100

# Read initial coordinates
file_path = os.path.join('lj_sample_configurations','lj_sample_config_periodic1.txt')
atomic_coordinates, box_length = mc.read_xyz(file_path)
```

```python
final_coordinates = mc.run_simulation(atomic_coordinates, box_length, cutoff, reduced_temperature, num_steps, max_displacement, freq)
```
